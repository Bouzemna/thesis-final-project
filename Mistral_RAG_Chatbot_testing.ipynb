```python
# Load your dataset
df = pd.read_csv("/content/expanded_qa_dataset (1).csv")  # Upload to Colab
df = df.dropna()
df = df[["question", "response"]]
```

```python
#  Embed questions using SentenceTransformers
embedder = SentenceTransformer('all-MiniLM-L6-v2', device='cpu')
corpus_embeddings = embedder.encode(df["question"].tolist(), convert_to_tensor=True).cpu()
```

# Load the Mistral Model and Tokenizer

```python
#  Load Mistral model from Hugging Face
model_name = "mistralai/Mistral-7B-Instruct-v0.1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", torch_dtype=torch.float16)
```

# Define the Hybrid RAG + Generation Function

```python
# Function to retrieve similar Q&A from dataset
def retrieve_similar_question(user_question, k=1):
    user_embedding = embedder.encode([user_question], convert_to_tensor=True)
    scores = cosine_similarity(user_embedding, corpus_embeddings)[0]
    top_k_idx = np.argsort(scores)[-k:][::-1]
    return df.iloc[top_k_idx[0]]
```

**Answer Generation Using Retrieval-Augmented Prompting with Mistral**

```python
def generate_answer_mistral(user_question):
    retrieved = retrieve_similar_question(user_question)

    context = f"Someone was once asked:\nQ: {retrieved['question']}\nA: {retrieved['response']}"
    prompt = (
        f"{context}\n\n"
        f"Now, based on that example, answer this new question in a clear, personal, and reflective way.\n"
        f"Q: {user_question}\nA:"
    )

    # Placer les inputs sur le même device que le modèle
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    # Génération
    with torch.no_grad():
        output = model.generate(**inputs, max_new_tokens=100)

    # C’est ici que l’erreur venait → il faut .cpu() avant decode
    decoded_output = tokenizer.decode(output[0].cpu(), skip_special_tokens=True)
    return decoded_output



```

```python
import time
from sklearn.metrics.pairwise import cosine_similarity
from bert_score import score as bert_score_fn  # pour éviter l'erreur 'module object is not callable'

# Échantillon à évaluer
test_df = df.sample(n=3, random_state=42).reset_index(drop=True)

results = []
start = time.time()

for i, row in test_df.iterrows():
    print(f"Processing question {i+1}/{len(test_df)}")
    question = row["question"]
    reference = row["response"]
    generated = generate_answer_mistral(question)

    # BLEU (texte brut uniquement)
    bleu_result = bleu.compute(
        predictions=[generated],
        references=[reference]
    )

    # ROUGE
    rouge_result = rouge.score(reference, generated)

    # BERTScore
    P, R, F1 = bert_score_fn([generated], [reference], lang="en", verbose=False)

    # Cosine Similarity
    ref_embed = embedder.encode([reference], convert_to_tensor=True).cpu()
    gen_embed = embedder.encode([generated], convert_to_tensor=True).cpu()
    cosine_sim = cosine_similarity(ref_embed.numpy(), gen_embed.numpy())[0][0]

    results.append({
        "Question": question,
        "Reference": reference,
        "Generated": generated,
        "BLEU": bleu_result["bleu"],
        "ROUGE-1": rouge_result["rouge1"].fmeasure,
        "ROUGE-L": rouge_result["rougeL"].fmeasure,
        "BERT_P": P[0].item(),
        "BERT_R": R[0].item(),
        "BERT_F1": F1[0].item(),
        "Cosine Similarity": cosine_sim
    })

end = time.time()
print(f"Evaluation completed in {round(end - start, 2)} seconds.")

# Résultats
results_df = pd.DataFrame(results)
results_df

# Export vers CSV
results_df.to_csv("mistral_rag_results.csv", index=False)
print("Résultats exportés dans 'mistral_rag_results.csv'")

```
