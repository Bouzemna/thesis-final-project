# Let's paraphrase AND expand our orginal dataset

```python
#  Installation des bibliothèques


# Charger ton dataset
df = pd.read_csv("/content/cleaned_responses (2) (2) (1).csv")
df = df.dropna(subset=["question", "response"])
df["question"] = df["question"].astype(str).str.strip()
df["response"] = df["response"].astype(str).str.strip()

#  Charger le modèle de paraphrasing
paraphrase_model_name = "ramsrigouthamg/t5_paraphraser"

tokenizer = AutoTokenizer.from_pretrained(paraphrase_model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(paraphrase_model_name).to("cuda" if torch.cuda.is_available() else "cpu")

#  Fonction pour paraphraser une question
def paraphrase_question(question, num_return_sequences=2):
    text = f"paraphrase: {question} </s>"
    encoding = tokenizer.encode_plus(
        text, padding="max_length", return_tensors="pt", max_length=128, truncation=True
    )
    input_ids = encoding["input_ids"].to(model.device)
    attention_mask = encoding["attention_mask"].to(model.device)

    outputs = model.generate(
        input_ids=input_ids,
        attention_mask=attention_mask,
        max_length=128,
        num_beams=5,
        num_return_sequences=num_return_sequences,
        early_stopping=True
    )

    return list(set([
        tokenizer.decode(output, skip_special_tokens=True, clean_up_tokenization_spaces=True)
        for output in outputs
    ]))

# Créer les nouvelles lignes
expanded_rows = []

for _, row in df.iterrows():
    question = row["question"]
    answer = row["response"]

    try:
        paraphrases = paraphrase_question(question, num_return_sequences=2)
        for para in paraphrases:
            if para.lower() != question.lower():
                expanded_rows.append({"question": para, "response": answer})
    except Exception as e:
        print(f"Erreur avec '{question}' → {e}")

#  Combiner original + paraphrases
df_paraphrased = pd.DataFrame(expanded_rows)
df_final = pd.concat([df, df_paraphrased]).drop_duplicates().reset_index(drop=True)

# Sauvegarder
df_final.to_csv("/content/expanded_qa_dataset.csv", index=False)
print(f" Fichier créé avec {len(df_final)} lignes")
```

```python


# 1. Load your original dataset
df = pd.read_csv("/content/expanded_qa_dataset.csv")
df = df.dropna(subset=["question", "response"])
df["question"] = df["question"].astype(str).str.strip()
df["response"] = df["response"].astype(str).str.strip()

# 2. Clean repeated phrases in responses
def clean_response(text):
    sentences = [s.strip() for s in re.split(r'[.!?]', text) if s.strip()]
    seen = set()
    unique_sentences = []
    for s in sentences:
        if s.lower() not in seen:
            seen.add(s.lower())
            unique_sentences.append(s)
    return ". ".join(unique_sentences).strip() + "."

# 3. Garbage detector (returns True if the text is bad)
def is_garbage(text):
    return (
        len(text.strip()) < 15 or
        bool(re.fullmatch(r"([A-Za-z]\.?)+", text.strip())) or
        len(set(text.strip())) <= 3
    )

# 4. Apply cleaning + filtering
df["cleaned_response"] = df["response"].apply(clean_response)
df = df[df["question"].apply(lambda x: not is_garbage(x))]
df = df[df["cleaned_response"].apply(lambda x: not is_garbage(x))]

# 5. Save cleaned file
df_final = df[["question", "cleaned_response"]].rename(columns={"cleaned_response": "response"})
df_final.to_csv("/content/expanded_qa_dataset_improved.csv", index=False)

print("Cleaned dataset saved as: expanded_qa_dataset_improved.csv")

```

# User interface to compare the three models

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Version 1: Small model (baseline)
small_tokenizer = T5Tokenizer.from_pretrained("./flan-t5-finetuned-qabot")
small_model = T5ForConditionalGeneration.from_pretrained("./flan-t5-finetuned-qabot").to(device)

# Version 2: Personalized model (v3, still experimental)
v3_tokenizer = T5Tokenizer.from_pretrained("./flan-t5-v3-personalized")
v3_model = T5ForConditionalGeneration.from_pretrained("./flan-t5-v3-personalized").to(device)

# Version 3: BEST fine-tuned model – Emna
emna_tokenizer = T5Tokenizer.from_pretrained("./flan-t5-base-finetuned-emna")
emna_model = T5ForConditionalGeneration.from_pretrained("./flan-t5-base-finetuned-emna").to(device)
```

```python
def generate_response(model, tokenizer, question):
    input_text = "You are warm and expressive. question: " + question
    input_ids = tokenizer.encode(input_text, return_tensors="pt").to(device)
    output = model.generate(
        input_ids,
        max_length=80,
        min_length=10,
        num_beams=4,
        no_repeat_ngram_size=3,
        early_stopping=True
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
def compare_models(question):
    output_small = generate_response(small_model, small_tokenizer, question)
    output_v3 = generate_response(v3_model, v3_tokenizer, question)
    output_emna = generate_response(emna_model, emna_tokenizer, question)
    return output_small, output_v3, output_emna

```

```python
iface = gr.Interface(
    fn=compare_models,
    inputs=gr.Textbox(lines=2, placeholder="Ask a question like 'What motivates you?'"),
    outputs=[
        gr.Textbox(label="Flan-T5 Small (Baseline)"),
        gr.Textbox(label="Flan-T5 V3 (Experimental)"),
        gr.Textbox(label="Flan-T5 Emna (Best Fine-Tuned )")
    ],
    title="Bot – 3-Model Comparison",
    description="Type a question and see how all 3 models answer it. Emna is your strongest and most natural version!"
)

iface.launch()

```
